## Here I train and create the model for Ada-boosting which will predict the "Inspection Result" Feature

```python
ins_x = ins.drop('Inspection Result', axis = 1)
ins_y = ins['Inspection Result']
X_train, X_test, y_train, y_test = train_test_split(ins_x, ins_y, shuffle = True, stratify=ins_y, test_size=0.20)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

cat_attribs = ['Inspection Date', 'Industry', 'Borough', 'Street', 'City']
#'Building Number', 'State'
full_pipeline = ColumnTransformer([
        #("num", num_pipeline, num_attribs),
        ("cat", OneHotEncoder(), cat_attribs),
    ])

X_train = full_pipeline.fit_transform(X_train)
X_test = full_pipeline.transform(X_test)
```

```python
y_train.value_counts() / len(y_train)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

param_grid = [
    {'learning_rate': [0.1,0.5,1,5,10,50,100]},
  ]
#learning_rate=0.5,
ada_clf = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1), n_estimators=200,
    algorithm="SAMME.R", random_state=42)

grid_search = GridSearchCV(ada_clf, param_grid=param_grid, cv=5,
                           scoring='accuracy', return_train_score=True, n_jobs = -1)
  

grid_search.fit(X_train, y_train)
```

```python
grid_search.best_estimator_
```

```python
#y_pred = ada_clf.predict(X_test)
y_pred = grid_search.predict(X_test)
```

```python
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import roc_curve
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

print("Accuracy: {:.2f}%".format(100 * accuracy_score(y_test, y_pred)))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
